## lris 붓꽃 데이터셋

```python
from sklearn.datasets import load_iris
iris = load_iris()
iris
```

# DataFrame 변환

```python
import pandas as pd
df = pd.DataFrame(data=iris['data'], columns=iris['feature_names'])
df.head()
```

```python
df['target'] = iris['target']
df[:100]
```

# 데이터셋 시각화
## Matplotlib 패키지
* 차트 또는 그림을 표시한다

## Seaborn 패키지
* 자주 사용하는 그래프를 편리하게 사용할 수 있게한다
* 예쁘게 그린다

## 꽃받침의 길이와 너비에 따른 라벨 분포
* Scatter Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(16,8))
sns.scatterplot(x=df['petal length (cm)'], y=df['petal width (cm)'], hue=df['target'])
plt.show()
```

## 라벨의 분포 그래프
* Distribution Plot

- 불균형 분포 그래프는 학습시키기 어렵다

```python
sns.displot(x=df['target'])
plt.show()
```

## 꽃받침 길이의 분포 그래프

```python
sns.displot(x=df['sepal length (cm)'])
plt.show()
```
